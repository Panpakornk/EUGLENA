```python
import os
from ultralytics import YOLO
%matplotlib inline
from matplotlib import pyplot as plt
from PIL import Image

import torch


#Define a project --> Destination directory for all results
project = "E:\\EUGLENA\\Results\\P\\1280"
#Define subdirectory for this specific training
name = "1280_batch4_100epochs" 

# Load a model
model = YOLO("yolo26n-seg.yaml")  # build a new model from YAML
model = YOLO("yolo26n-seg.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolo26n-seg.yaml").load("yolo26n.pt")  # build from YAML and transfer weights

# Train the model
results = model.train(data='E:\EUGLENA\Dataset\P\P.v1-p_1280.yolo26\data.yaml',
                      project=project,
                      name=name,
                      epochs=100,
                      patience=0, #setting patience=0 to disable early stopping.
                      batch=4,
                      imgsz=1280,  
                    )
```
